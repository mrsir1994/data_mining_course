```python
# define the stemmer
import nltk

analyzer = CountVectorizer().build_analyzer()
def stemmed_words(doc):
    stemmer = nltk.stem.SnowballStemmer('english')
    return (stemmer.stem(w) for w in analyzer(doc))
```

The lemmatizer above is defined to be used to lemmatize the dataset terms. A set of stopwords united from NLTK's stopwords and sklearn's stopwords and string's punciation list, is used to exclude these stopwords from the features. 

```python
# test code
corpus = ['my name is yes', 'what is your name?', 'do you like music music music', 'i hate you']

vectorizer1 = CountVectorizer(stop_words = combined_stopwords,min_df = 2,analyzer=stemmed_words)
vectorizer2 = CountVectorizer(stop_words = combined_stopwords,min_df = 5,analyzer=stemmed_words)
data_vec1 = vectorizer1.fit_transform(train.data)
data_vec2 = vectorizer2.fit_transform(train.data)

print 'num of term extracted when min_df = %d is %d' %(2, data_vec1.shape[1])
print 'num of term extracted when min_df = %d is %d' %(5, data_vec2.shape[1])

data_vect = data_vec2

tfTrans = TfidfTransformer()
tf_vec = tfTrans.fit_transform(data_vect)
```

```python
def plot_roc(fpr, tpr):
    fig, ax = plt.subplots()

    roc_auc = auc(fpr,tpr)

    ax.plot(fpr, tpr, lw=2, label= 'area under curve = %0.4f' % roc_auc)

    ax.grid(color='0.7', linestyle='--', linewidth=1)

    ax.set_xlim([-0.1, 1.1])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate',fontsize=15)
    ax.set_ylabel('True Positive Rate',fontsize=15)

    ax.legend(loc="lower right")

    for label in ax.get_xticklabels()+ax.get_yticklabels():
        label.set_fontsize(15)

def fit_predict_and_plot_roc(pipe, train_data, train_label, test_data, test_label):
    pipe.fit(train_data, train_label)
    # pipeline1.predict(twenty_test.data)

    prob_score = pipe.decision_function(test_data)
    pred_labels = pipe.predict(test_data)
    fpr, tpr, _ = roc_curve(test_label, prob_score)

    plot_roc(fpr, tpr)
    print 'confusion matrix:'
    print confusion_matrix(test_label,pred_labels)
    print '-'*20
    print 'accuracy:'
    print accuracy_score(test_label, pred_labels)
    print '-'*20
    print 'precision:'
    print precision_score(test_label, pred_labels,average='macro')
    print '-'*20
    print 'recall:'
    print recall_score(test_label, pred_labels, average='macro')
    print '-'*20
```
